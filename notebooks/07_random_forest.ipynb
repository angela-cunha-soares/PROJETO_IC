# 07 · Random Forest

Rotulagem CONAMA → **severidade** (alvo balanceado), anti-leakage, métricas e importância de variáveis.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent / 'src'))
import matplotlib.pyplot as plt
import pandas as pd
import config
from features.build_features import rotular_severidade, rotular_conama, features_sem_vazamento, contar_violacoes
from models import random_forest as rf
scaled = pd.read_csv(config.PROCESSED_DIR / 'dados_scaled.csv')
imp = pd.read_csv(config.PROCESSED_CSV)
ID = ['data','Ano','Mes','mes_num','trimestre','estacao']
feats = [c for c in scaled.columns if c not in ID]
```

```python
print('rótulo estrito (degenerado):', rotular_conama(imp)[0].value_counts().to_dict())
y, usadas = rotular_severidade(imp)
print('severidade:', y.value_counts().to_dict())
Xfeats = features_sem_vazamento(feats, usadas)
print('features RF (sem vazamento):', Xfeats)
```

```python
res = rf.treinar(scaled[Xfeats], y)
print(res.metricas, '| CV-F1:', round(res.cv_f1.mean(),3))
from visualization import plots
plots.matriz_confusao(res.matriz_confusao); plt.show()
plots.importancia_variaveis(res.importancias); plt.show()
```
