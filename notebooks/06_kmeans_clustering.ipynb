Cotovelo + silhueta, ajuste e perfil dos clusters.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent / 'src'))
import matplotlib.pyplot as plt
import pandas as pd
import config
from models import kmeans as km
scaled = pd.read_csv(config.PROCESSED_DIR / 'dados_scaled.csv')
imp = pd.read_csv(config.PROCESSED_CSV)
ID = ['data','Ano','Mes','mes_num','trimestre','estacao']
feats = [c for c in scaled.columns if c not in ID]
X = scaled[feats].to_numpy(float)
```
